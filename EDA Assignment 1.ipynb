Importing the pandas library.
Reading the CSV file named 'housing.csv' into a DataFrame.

```python
import pandas as pd

df = pd.read_csv('housing.csv')
```

```python
# Just taking a look on the how the dataset is look like

df
```

```python
# We find the summary of the dataset
# We get to know about the dataset statistic like the count, mean, std, min and quartiles

df.describe()
```

```python
# Importing the Seaborn(data visualization library)
import seaborn as sns

# We try to generate a scatter plot of the dataset
# The scatter plot visualizes the geographical locations of areas in California based on their longitude and latitude coordinates.
# Each point on the plot represents a specific geographical location.
# Clustering of points may indicate densely populated or urban areas.
# Sparse areas with fewer data points may represent rural or less densely populated regions.
g = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude')
g.set(title='Geographical Location of Areas in California')
```

```python
g = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='median_house_value')
g.set(title='Geographical Location of Areas in California')

# in this plot, the color of each point provides additional information about the median house value in that area.
```

```python
g = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='ocean_proximity')
g.set(title='Geographical Location of Areas in California')

# Each point is color-coded based on the category of ocean proximity for that area (e.g., 'NEAR BAY', 'NEAR OCEAN', etc.).
# This color encoding helps identify areas located near different bodies of water, such as bays or oceans.
```

```python
# The new column 'above_median_house_value' provides a binary classification of whether a house value is above or below the median value in the dataset.
# This binary attribute can be useful for classification tasks or for segmenting the dataset based on house values relative to the median.


df['above_median_house_value'] = (df.median_house_value > df.median_house_value.median()).astype(int)
df
```

```python
g = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='above_median_house_value')

# Each point is color-coded based on whether the house value for that area is above (different color) or below (same color) the median value.
# This binary color encoding allows for the identification of areas with house values above or below the median, providing insights into spatial patterns of housing affordability.
```

```python
# Each point is styled based on whether the house value for that area is above (different marker style) or below (same marker style) the median value.

g = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', style='above_median_house_value')
```

```python
# This binary marker styling and color encoding provide redundant visual cues for identifying areas with house values above or below the median, facilitating easier interpretation of the plot.

g = sns.relplot(kind='scatter', data=df, x='longitude',
                y='latitude', style='above_median_house_value',
                hue='above_median_house_value')
```

```python
# Marker style and color indicate whether the house value for each area is above or below the median, providing redundant visual cues.
# Marker size provides additional information about the median house value, allowing for a comparison of relative values across different areas.

g = sns.scatterplot(data=df, x='longitude',
                y='latitude', style='above_median_house_value',
                hue='above_median_house_value',
                size='median_house_value')
```

```python
# It visualizes the relationship between the 'total_rooms' attribute (representing the total number of rooms in a housing unit) and the 'ocean_proximity' attribute

sns.catplot(x='total_rooms', y='ocean_proximity', data=df)
```

```python
#  It uses the 'above_median_house_value' attribute as a hue to distinguish between house values above and below the median.
sns.catplot(x='median_house_value', y='ocean_proximity', data=df, hue='above_median_house_value')
```

```python
# This visualization allows for a comparison of the distribution of median house values across different categories of ocean proximity. It helps in understanding the range and variability of house values within each category and can identify potential outliers or differences in house values based on their proximity to the ocean.

sns.boxplot(x='median_house_value', y='ocean_proximity', data=df)
```

```python
# It helps in understanding how housing age varies based on proximity to the ocean and whether houses fall above or below the median value.

sns.boxplot(x='housing_median_age', y='ocean_proximity',
            hue='above_median_house_value', data=df)
```

```python
# This visualization provides a detailed view of the distribution of median house values across different categories of ocean proximity. It allows for an understanding of the variability and shape of the distributions, as well as any potential outliers or differences in house values based on their proximity to the ocean.

sns.violinplot(y="median_house_value", x="ocean_proximity", data=df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant numerical attributes for pair plot
numerical_attributes = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']

# Create pair plot
sns.pairplot(df[numerical_attributes])
plt.show()

# the pair plot provides a comprehensive visual overview of the relationships between different numerical attributes in the dataset, helping to identify patterns, correlations, and potential outliers
```

```python
# Drop non-numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()


# From this we will be able to find that there is a relation between the income and the house value i.e. the people with high income buy a expensive house and vice-varsa.
```

```python
numerical_attributes = ['housing_median_age', 'median_income', 'median_house_value']

# Set up subplots
fig, axes = plt.subplots(nrows=len(numerical_attributes), ncols=1, figsize=(8, 6 * len(numerical_attributes)))

# Plot histograms and KDE plots for each numerical attribute
for i, attribute in enumerate(numerical_attributes):
    sns.histplot(df[attribute], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'{attribute} Distribution', fontsize=14)
    axes[i].set_xlabel(attribute, fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()
# We can easliy find that the housing age less than has very less frequecy while the mostly the age of the house is between 15 to 35 years.
# From these we can also see that the pattern of the median salary and the median house is somewhat similar
```
